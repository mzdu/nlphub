```python
# Create your first MLP in Keras
from keras.models import Sequential
from keras.layers import Dense
import numpy
```

```python
# fix random seed for reproducibility
# use a stochastic process
seed = 7
numpy.random.seed(seed)



# load pima indians dataset
dataset = numpy.loadtxt("pima-indians-diabetes.csv", delimiter=",")
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]





# create model
# we create a sequential model and add layers one at a time until we are heppy with our network topology.
model = Sequential()


# Dense means fully connected
# input_dim is input dimension
# relu, rectifier
# sigmoid, output 0~1

# first hidden layer, has 12 neurons, 8 input variables
model.add(Dense(12, input_dim=8, init='uniform', activation='relu'))
# second hidden layer has 8 neurons
model.add(Dense(8, init='uniform', activation='relu'))
# output layer has 1 neuron to predict the class
model.add(Dense(1, init='uniform', activation='sigmoi'))
```

```python
# Compile model

# adam, stochastic optimizer
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])


# Fit the model / train our model
# nb_epoch, the training process runs for a fixed number of iterations through the dataset called epochs.
# batch_size, the number of instances that are evaluated before a weight update in the network.
model.fit(X, Y, nb_epoch=150, batch_size=10)




# evaluate the model
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
