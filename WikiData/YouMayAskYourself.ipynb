```python
# You May Ask Yourself Parser
from bs4 import BeautifulSoup
soup = BeautifulSoup(open("YouMayAskYourself.xhtml"), 'html.parser')


terms = soup.findAll('span', {'class':'color'})
defs = soup.findAll('p', {'class':'noindent1'})

# print terms

file = open('YouMayAskYourself.txt', 'w')


i = 0
while i < len(defs):
    term = terms[i].get_text()
    
    
    definition = defs[i].get_text()[len(term):]


    print term + ":" + definition
    
    file.write(term.encode('utf-8') + " : ")
    file.write(definition.encode('utf-8') + '\n')
    
    i += 1

file.close()
```
