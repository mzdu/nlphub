```python
# Sociology13 Parser
from bs4 import BeautifulSoup
soup = BeautifulSoup(open("Sociology13.html"), 'html.parser')


terms = soup.findAll('span', {'class':'cell-word ng-binding'})
defs = soup.findAll('span', {'class':'cell-meaning ng-binding'})
combos = []

file = open('sociology13.txt', 'w')


i = 0
while i < len(terms):
    term = terms[i].get_text()
    definition = defs[i].get_text()
    print term + ' : ' + definition
    
    file.write(term.encode('utf-8') + " : ")
    file.write(definition.encode('utf-8') + '\n')
    
    i += 1

file.close()
```
